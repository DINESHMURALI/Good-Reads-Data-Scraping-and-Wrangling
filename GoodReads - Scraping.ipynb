```python
import pandas as pd
import bs4
import selenium
import requests
```

```python
from selenium import webdriver
browser = webdriver.Chrome('chromedriver.exe')
browser.get('https://www.goodreads.com/choiceawards/best-books-2017')
```

```python
books = browser.find_elements_by_css_selector('div.category.clearFix')
print(len(books))
```

```python
df_books = pd.DataFrame(columns=['Category','Book Name','Image','URL'])

for cat in books:
    
    category = cat.find_element_by_css_selector('h4')
    category = category.get_attribute('innerHTML')
    
    name = cat.find_element_by_css_selector('img.winner')
    name = name.get_attribute('alt')
    
    url = cat.find_element_by_css_selector('img.winner')
    url = url.get_attribute('src')
    
    url_book = cat.find_element_by_css_selector('a')
    url_book = url_book.get_attribute('href')

    curr_row = {
        'Category':category,
        'Book Name': name,
        'Image': url,
        'URL': url_book
    }
    
    df_books = df_books.append(curr_row, ignore_index=True)
df_books
    
    #print(category,'|',name,'|', url,'|', url_book)
```

```python
df_books.to_csv('books_scrapping.csv', index = False)
```

```python
df_books_1 = pd.DataFrame(columns=['Book Name','Author','Description','Votes','Total Votes'])

for url in df_books['URL']:
    browser.get(url)
    
    title = browser.find_element_by_css_selector('a.winningTitle')
    title = title.get_attribute('innerHTML')
    
    p_full_aname = browser.find_element_by_css_selector('a.authorName')
    c_full_aname = p_full_aname.find_element_by_css_selector('span[itemprop]')
    aname = c_full_aname.get_attribute('innerHTML')
    
    desc = browser.find_element_by_css_selector('div.readable.stacked')
    desc = desc.get_attribute('innerHTML')
    
    votes = browser.find_element_by_css_selector('span.greyText')
    votes = votes.get_attribute('innerHTML')
    
    tot_votes = browser.find_element_by_css_selector('greytext.gcaNumVotes')
    tot_votes = tot_votes.get_attribute('innerHTML')
    
    
    curr_row = {
        'Book Name': title,
        'Author': aname,
        'Description': desc,
        'Votes': votes,
        'Total Votes': tot_votes
    }
    
    df_books_1 = df_books_1.append(curr_row, ignore_index = True)
df_books_1
```

```python
df = pd.merge(df_books,df_books_1)
```
